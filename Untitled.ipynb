```python
# 1. Проверяем, что мы в нужной папке
pwd

# 2. Инициализируем git
git init

# 3. Добавляем все файлы проекта
git add .

# 4. Делаем первый коммит
git commit -m "Initial commit: HW_03 Python DS 2025"

# 5. Подключаем новый репозиторий GitHub
# (вместо твоего_юзернейма укажи свой GitHub username)
git remote add origin https://github.com/твой_юзернейм/HW_03_python_ds_2025.git

# 6. Отправляем файлы на GitHub
git branch -M main
git push -u origin main
```
